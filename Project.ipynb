```python
import pandas as pd
df=pd.read_csv('dtrump_cleaneddata.csv')
df
```

```python
df[pd.to_numeric(df.tweet, errors='coerce').isnull()]
```

```python
#Removing special characters and punctuations
df['Special Characters removal']=df['tweet'].str.replace('[!-;:,<>./?@#$%^&*_~{}]','')
#Remove numbers
df['Remove numbers']=df['tweet'].str.replace('\d','')
#Stopwords removal
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english') 
df['Stopwords removal']=df['tweet'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
#Lower casing
df['Lower casing']=df['tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
#stemming
from nltk.stem import PorterStemmer
st=PorterStemmer()
df['stemming']=df['tweet'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
#Lemmatization
import nltk
from textblob import Word
nltk.download('wordnet')
df['Lemmatization']=df['tweet'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```
